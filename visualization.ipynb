```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_json("items.json")
data
```

```python
data = data.drop(columns=['book_link'])
data
```

```python
df = pd.DataFrame({
    'book_name': [j for i in data['book_name'] for j in i],
    'book_author': [j for i in data['book_author'] for j in i],
    'book_score': [j for i in data['book_score'] for j in i],
    'book_vote': [j for i in data['book_vote'] for j in i],
    'book_ratings': [j for i in data['book_ratings'] for j in i]
})

df
```

```python
df[['avg_book_rating', 'book_ratings']] = df['book_ratings'].apply(lambda x: pd.Series(x.split('— ')))
df
```

```python
df['book_score'] = df['book_score'].replace({'score:':''}, regex=True)
df['book_vote'] = df['book_vote'].replace({'people voted':''}, regex=True)
df['book_ratings'] = df['book_ratings'].replace({'ratings':'','rating':''}, regex=True)
df['avg_book_rating'] = df['avg_book_rating'].replace({'avg rating':''}, regex=True)

df
```

```python
df.describe()
```

```python
df.info()
```

```python
df['book_score'] = df['book_score'].str.replace(',', '').astype(int)
df['book_vote'] = df['book_vote'].str.replace(',', '').astype(int)
df['book_ratings'] = df['book_ratings'].str.replace(',', '').astype(int)
df['avg_book_rating'] = df['avg_book_rating'].str.replace(',', '').astype(float)

df.info()
```

```python
df = df.drop_duplicates()
```

```python
totalScoreOfAuthors = df.groupby('book_author')['book_score'].sum().reset_index()
totalScoreOfAuthors.columns = ['book_author','book_score']

top_5_authors_by_score = totalScoreOfAuthors.sort_values(by='book_score',ascending=False).head(5)
remaining_authors_score = totalScoreOfAuthors.sort_values(by='book_score',ascending=False)[5:]
remaining_score = remaining_authors_score['book_score'].sum()

other_row = pd.DataFrame({'book_author':['others'],'book_score':[remaining_score]})

top_5_authors_by_score = pd.concat([top_5_authors_by_score,other_row],ignore_index=True)

plt.figure()
plt.pie(top_5_authors_by_score['book_score'],labels=top_5_authors_by_score['book_author'],autopct='%1.1f%%')
plt.title('Top 5 authors by score')
plt.show()
```

```python
totalVotesOfAuthors = df.groupby('book_author')['book_vote'].sum().reset_index()
totalScoreOfAuthors.columns = ['book_author','book_vote']

top_10_authors_by_votes = totalScoreOfAuthors.sort_values(by='book_vote',ascending=False).head(10)


plt.figure()
sns.barplot(x=top_10_authors_by_votes['book_vote'],y=top_10_authors_by_votes['book_author'])
plt.title('Top 10 authors by votes')
plt.show()
```

```python
top_10_books = df[['book_name','book_score','book_vote']].sort_values(by='book_score',ascending=False).head(10)

plt.figure(figsize=(10,10))
sns.scatterplot(data=top_10_books, x='book_score', y='book_vote', hue='book_name', style='book_name', s=100)
plt.show()
```

```python
plt.figure()
top_10_books_by_rating = df[['book_name','book_ratings']].sort_values(by='book_ratings',ascending=False)[1:10]

sns.barplot(data=top_10_books_by_rating,x='book_ratings',y='book_name')
plt.show()
```
